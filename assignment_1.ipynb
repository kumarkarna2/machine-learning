```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('dataset/transistors.csv')

x = data.iloc[:,0]
y = data.iloc[:,1]

#non_linear relationship
plt.plot(x,y,'o')
plt.xlabel('Year')
plt.ylabel('Transistors')
plt.show()


#log transformation and linear relationship
y=np.log(y)

denominator = x.dot(x) - x.mean()*x.sum()
m=(x.dot(y)-y.mean()*x.sum())/denominator
c = (y.mean()*x.dot(x) - x.mean()*x.dot(y))/denominator
line = m*x + c


plt.scatter(x,y)
plt.plot(x,line,color='red')
plt.xlabel('Year')
plt.ylabel('Transistors count ofter log transformation')
plt.show()
```

```python
from matplotlib.figure import Figure

data = pd.read_csv('dataset/city_day.csv')
delhi = data[data['City']=='Delhi']
aqi = delhi['AQI']
aqi.fillna(aqi.mean())

date = delhi['Date']
date=np.array(date)
date = pd.to_datetime(date)



plt.figure(figsize=(25,5))
plt.plot(date,aqi)
plt.title('Air Quality Index of Delhi from 2015-2020')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.show()


```

carbon monoxide

```python
co = delhi['CO']
co.fillna(co.mean())

co_before_2020 = delhi[delhi['Date']<'2020-01-01'].mean(numeric_only=True)
co_after_2020 = delhi[delhi['Date']>='2020-01-01'].mean(numeric_only=True)


plt.bar('Before 2020',co_before_2020,color='r')
plt.bar('After 2020',co_after_2020,color='g')
plt.title('Average Carbon monoxide before and after 2020')
plt.show()
```

PM2.5

```python
pm2_5 = delhi['PM2.5']
pm2_5.fillna(pm2_5.mean(numeric_only=True))

plt.figure(figsize=(25,5))
plt.plot(date,pm2_5)
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.title('PM 2.5 of Delhi from 2015-2020')
plt.show()
```

Ozone

```python
import seaborn as sns

o3 = delhi['O3']
o3.fillna(o3.median(),inplace=True)

# o3.isnull().sum()

plt.figure(figsize=(15,5))
plt.plot(date,o3,color='r')
plt.xlabel('Date')
plt.ylabel('O3')
plt.title('Ozone of Delhi from 2015-2020')
plt.show()
```
